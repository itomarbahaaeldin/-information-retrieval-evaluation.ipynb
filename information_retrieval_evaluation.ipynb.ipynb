```python
# Import necessary libraries
import pandas as pd
import spacy
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity

```

```python
# Load the English language model for spaCy
nlp = spacy.load("en_core_web_sm")


```

```python
# Load the dataset from a CSV file
dataset = pd.read_csv("NLP_2.csv")
```

```python
# Extract the "ABSTRACT" column from the dataset to create the corpus
corpus = dataset["ABSTRACT"]
```

```python
# Define a function to preprocess the text
def preprocess_text(text):
    # Process the text with spaCy to create a Doc object
    doc = nlp(text)
    # Lemmatize tokens and remove stop words and punctuation
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    # Join the tokens back into a single string
    return " ".join(tokens)
```

```python
# Apply the preprocessing function to the entire corpus
corpus_cleaned = corpus.apply(preprocess_text)
```

```python
# Create a binary bag-of-words (BBoW) representation of the cleaned corpus
binary_vectorizer = CountVectorizer(binary=True)
BBoW = binary_vectorizer.fit_transform(corpus_cleaned)
```

```python
# Create a frequency bag-of-words (FBoW) representation of the cleaned corpus
frequency_vectorizer = CountVectorizer()
FBoW = frequency_vectorizer.fit_transform(corpus_cleaned)
```

```python
# Create TF-IDF representation again in case it was not defined
tfidf_vectorizer = TfidfVectorizer()
TFidf = tfidf_vectorizer.fit_transform(corpus_cleaned)

# Now calculate vocabulary sizes
BBoW_vocab_size = len(binary_vectorizer.vocabulary_)
FBoW_vocab_size = len(frequency_vectorizer.vocabulary_)
TFidf_vocab_size = len(tfidf_vectorizer.vocabulary_)

# Print the vocabulary sizes
print(f"BBoW Vocabulary Size: {BBoW_vocab_size}")
print(f"FBoW Vocabulary Size: {FBoW_vocab_size}")
print(f"TFidf Vocabulary Size: {TFidf_vocab_size}")

```

```python
# Get the index of the abstracts with ID 30 and ID 100
index_30 = dataset[dataset["id"] == 30].index[0]
index_100 = dataset[dataset["id"] == 100].index[0]
```

```python
# Extract the TF-IDF vectors for the abstracts with ID 30 and ID 100
tfidf_vector_30 = TFidf[index_30]
tfidf_vector_100 = TFidf[index_100]
```

```python
# Calculate the Euclidean distance between the two TF-IDF vectors
euclidean_dist = euclidean_distances(tfidf_vector_30, tfidf_vector_100)[0][0]
# Calculate the cosine similarity between the two TF-IDF vectors
cosine_sim = cosine_similarity(tfidf_vector_30, tfidf_vector_100)[0][0]

# Print the calculated distances and similarities
print(f"Euclidean Distance between ID 30 and ID 100: {euclidean_dist}")
print(f"Cosine Similarity between ID 30 and ID 100: {cosine_sim}")
```

```python
# Define a paragraph for which to find the nearest abstract
paragraph = "Unlocking the secrets of the universe requires unraveling its mathematical code, where data-driven insights guide our understanding of complex phenomena. Information, the lifeblood of our digital world, flows through algorithms, transforming bits into knowledge, all under the watchful eye of technology."
```

```python
# Preprocess the paragraph using the defined function
paragraph_cleaned = preprocess_text(paragraph)
```

```python
# Transform the cleaned paragraph into a TF-IDF vector
paragraph_tfidf = tfidf_vectorizer.transform([paragraph_cleaned])

```

```python
# Calculate cosine similarities between the paragraph vector and the TF-IDF matrix of the corpus
cosine_similarities = cosine_similarity(paragraph_tfidf, TFidf).flatten()
```

```python
# Find the index of the highest cosine similarity
nearest_index = cosine_similarities.argmax()
# Retrieve the nearest abstract and its ID based on the highest similarity
nearest_abstract = corpus.iloc[nearest_index]
nearest_id = dataset["id"].iloc[nearest_index]

# Print the ID and content of the nearest abstract
print(f"Nearest Abstract ID: {nearest_id}")
print(f"Nearest Abstract Content: {nearest_abstract}")
```

```python
# Retrieve and display the labels associated with the nearest abstract, excluding 'id' and 'ABSTRACT'
nearest_labels = (
    dataset.loc[dataset["id"] == nearest_id].drop(["id", "ABSTRACT"], axis=1).iloc[0]
)
print("Labels for the Nearest Abstract:")
print(nearest_labels)
```
